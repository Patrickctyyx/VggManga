```python
import os
import shutil
from utils import mkdir_if_not_exist
```

```python
img_dir = '../manga109'
for subdir in ['train', 'test']:
    dir_name = os.path.join(img_dir, subdir)
    imgs = os.listdir(dir_name)
    
    for img in imgs:
        if not img.endswith('.jpg'):
            continue
        img_sub_name = img.split('_')[0]
        mkdir_if_not_exist(os.path.join(dir_name, img_sub_name))
        shutil.move(os.path.join(dir_name, img), os.path.join(dir_name, img_sub_name, img))
```

```python
dir_name = os.path.join(img_dir, 'train')
val_dir = os.path.join(img_dir, 'validation')
sub_dirs = os.listdir(dir_name)

for sub_dir in sub_dirs:
    if sub_dir.startswith('.'):
        continue
    imgs = os.listdir(os.path.join(dir_name, sub_dir))
    for img in imgs:
        if not img.endswith('.jpg'):
            continue
        img_sub_name, img_num = img.split('_')
        img_num = int(img_num.split('.')[0])

        if img_num % 5 == 1:
            mkdir_if_not_exist(os.path.join(val_dir, img_sub_name))
            shutil.move(os.path.join(dir_name, img_sub_name, img), os.path.join(val_dir, img_sub_name, img))
```
